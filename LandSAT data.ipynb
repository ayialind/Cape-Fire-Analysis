```python
import os
import earthaccess
import pandas as pd
import geopandas as gp
import xarray as xr
```

```python
earthaccess.login(persist=True)
```

```python
Query = earthaccess.collection_query().keyword('landsat').provider('LPCLOUD')
print(f'Collections found: {Query.hits()}')

```

```python
geojson = gp.read_file('../Cape Fire Analysis/Cape_Floristic_Region.geojson')
geojson.geometry
```

```python
bbox = tuple(list(geojson.total_bounds))
bbox
```

```python
results = earthaccess.search_data(
    short_name='HLSL30',
    bounding_box=bbox,
    temporal=('2023-12-01','2023-12-31'),
    count = 584,
    cloud_cover = (0,10)
)
    
```

```python
pd.json_normalize(results)
```

```python
hls30_results_urls = [granule.data_links() for granule in results]
hls30_results_urls
```
